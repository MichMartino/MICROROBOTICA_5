```python
class Contatore:
    def __init__(self):
        self.punteggio = 0
        self.punteggioPiuAlto = 0

    def aggiungi_punti(self, punti):
        if punti > 0:
            self.punteggio += punti
        else:
            print("Disastro! No punteggio negativo")
        if self.punteggio > self.punteggioPiuAlto:
            self.punteggioPiuAlto = self.punteggio
        
    def getPunteggioAttuale(self):
        return self.punteggio

    def getPunteggioMassimo(self):
        return self.punteggioPiuAlto

contatore = Contatore()
contatore.aggiungi_punti(10)
contatore.aggiungi_punti(5)
print(f"Punteggio: {contatore.getPunteggioAttuale()}")
print(f"Record: {contatore.getPunteggioMassimo()}")
```
