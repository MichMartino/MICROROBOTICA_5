```python
#Alcuni telefoni rimossi, abbiamo dati puliti, 
#ora prepariamoli per la regressione

#LabelEncoder
labelEncoder_brand = LabelEncoder()
labelEncoder_model = LabelEncoder()
labelEncoder_store = LabelEncoder()
labelEncoder_display_type = LabelEncoder()
labelEncoder_display_resolution = LabelEncoder()
labelEncoder_os = LabelEncoder()
labelEncoder_usb = LabelEncoder()
labelEncoder_features_sensors = LabelEncoder()
labelEncoder_video = LabelEncoder()
labelEncoder_chipset = LabelEncoder()
labelEncoder_cpu = LabelEncoder()
labelEncoder_gpu = LabelEncoder()
labelEncoder_chip_company = LabelEncoder()
labelEncoder_gpu_company = LabelEncoder()
labelEncoder_fingerprint = LabelEncoder()

#Trasformo i dati per farli utilizzare alla regressione
dati_selezionati['brand_n'] = labelEncoder_brand.fit_transform(dati_selezionati["phone_brand"])
dati_selezionati['model_n'] = labelEncoder_model.fit_transform(dati_selezionati["phone_model"])
dati_selezionati['store_n'] = labelEncoder_store.fit_transform(dati_selezionati["store"])
dati_selezionati['display_type_n'] = labelEncoder_display_type.fit_transform(dati_selezionati["display_type"])
dati_selezionati['display_resolution_n'] = labelEncoder_display_resolution.fit_transform(dati_selezionati["display_resolution"])
dati_selezionati['os_n'] = labelEncoder_os.fit_transform(dati_selezionati["os"])
dati_selezionati['usb_n'] = labelEncoder_usb.fit_transform(dati_selezionati["usb"])
dati_selezionati['features_sensors_n'] = labelEncoder_features_sensors.fit_transform(dati_selezionati["features_sensors"])
dati_selezionati['video_n'] = labelEncoder_video.fit_transform(dati_selezionati["video"])
dati_selezionati['chipset_n '] = labelEncoder_chipset.fit_transform(dati_selezionati["chipset"])
dati_selezionati['cpu_n'] = labelEncoder_cpu.fit_transform(dati_selezionati["cpu"])
dati_selezionati['gpu_n'] = labelEncoder_gpu.fit_transform(dati_selezionati["gpu"])
dati_selezionati['chip_company_n'] = labelEncoder_chip_company.fit_transform(dati_selezionati["chip_company"])
dati_selezionati['gpu_company_n'] = labelEncoder_gpu_company.fit_transform(dati_selezionati["gpu_company"])
dati_selezionati['fingerprint_n'] = labelEncoder_fingerprint.fit_transform(dati_selezionati["fingerprint"])


```

```python
feature = ['storage', 'ram', 'weight',
       'nfc', 'battery', 'foldable', 'ppi_density', 'brand_n', 'model_n', 
           'store_n', 'display_type_n',
       'display_resolution_n', 'os_n', 'usb_n', 'features_sensors_n',
       'video_n', 'chipset_n ', 'cpu_n', 'gpu_n', 'chip_company_n',
       'gpu_company_n', 'fingerprint_n']
output = ['price_usd']

selected_data = dati_selezionati[feature+output]
```

```python
selected_data
```

```python
X = selected_data[feature].to_numpy()
y = selected_data[output].to_numpy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.3)
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train) #addestra le regressione lineare
```

```python
regressor.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Predizioni sul test set
y_pred = regressor.predict(X_test)

# Metriche di valutazione
mae  = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred) ** 0.5
r2   = r2_score(y_test, y_pred)

print(f"MAE  (errore medio assoluto) : {mae:.2f} $")
print(f"RMSE (scarto quadratico medio): {rmse:.2f} $")
print(f"R²   (coefficiente di det.)  : {r2:.4f}")

# Confronto visivo: valori reali vs predetti
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.4, edgecolors="k", linewidths=0.3)
min_val = min(y_test.min(), y_pred.min())
max_val = max(y_test.max(), y_pred.max())
plt.plot([min_val, max_val], [min_val, max_val], "r--", label="Predizione perfetta")
plt.xlabel("Prezzo reale (USD)")
plt.ylabel("Prezzo predetto (USD)")
plt.title("Reale vs Predetto - Regressione Lineare")
plt.legend()
plt.tight_layout()
plt.show()

# Distribuzione dei residui
residui = y_test.flatten() - y_pred.flatten()
plt.figure(figsize=(8, 4))
plt.hist(residui, bins=40, color="steelblue", edgecolor="white")
plt.axvline(0, color="red", linestyle="--")
plt.xlabel("Residuo (reale - predetto)")
plt.ylabel("Frequenza")
plt.title("Distribuzione dei residui")
plt.tight_layout()
plt.show()
```
