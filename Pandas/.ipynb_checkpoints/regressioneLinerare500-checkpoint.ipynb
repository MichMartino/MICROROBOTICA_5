## REGRESSIONI LINEARI

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("Used_fiat_500_in_Italy_dataset.csv")
data.head()
```

```python
data.info()
```

```python
feature = ["km"]
output = ["price"]
selected_data = data[feature+output]
selected_data.head()
```

```python
x = selected_data[feature].to_numpy()
y = selected_data[output].to_numpy()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.3)
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x_train, y_train) #addestra le regressione lineare
```

```python
a = regressor.coef_[0] #Coefficiente angolare
```

```python
b = regressor.intercept_ #termine noto
```

```python
ascissa = np.linspace(0, 200_000, 2) #per fare una retta bastano 2 punti
ordinata = a * ascissa + b

fig, ax = plt.subplots(figsize=(4,4))
ax.scatter(x, y, s = 10)
ax.plot(ascissa, ordinata, color = "red")
ax.set_xlabel("km")
ax.set_ylabel("price")
plt.show()
```

```python
regressor.predict([[10_000]])
```

## Test per vedere se la regressione funziona bene

La chiamata regressor.score(x_test, y_test) restituisce il coefficiente di determinazione $R^2$

- la regressione è buona tanto più $R^2$ si avvicina a 1;
- valori di $R^2$ inferiori a 0.5 indicano che la regressione non rispecchia i dati

```python
# La regressione funziona bene?
regressor.score(x_test, y_test)
```
