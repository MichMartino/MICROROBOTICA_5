```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def image2imageShaped(image):
    img_numero = np.reshape(image, (28, 28))
    return img_numero
```

```python
#100 assi, come l'esercizio di prima, 10 cifre a caso, un'unico titolo per ogni riga. 10 x 10
```

```python
data = pd.read_csv("./mnist_784_csv.csv")

columns_names = list(data.columns)[:-1]
#images = data[columns_names].to_numpy().astype(np.uint8)

pixel_dict = {'pixel1': '0', 'pixel2': '1', 'pixel3': '2', 'pixel4': '3', 'pixel5': '4','pixel6': '5', 'pixel7': '6', 'pixel8': '7', 'pixel9': '8', 'pixel10': '9'}
```

```python
fig, axs = plt.subplots(nrows=10, ncols=10, figsize = (18, 18)) #axs è una matrice di numpy 10 x 10
np.random.seed(42)
for number, ax in enumerate(axs): #Scorre sulle righe, ax è un array di 10
    selected = data[data["class"] == number][columns_names] #Select pixel1, pixel2, pixel... FROM data WHERE data.class = number 
    array = selected.to_numpy()
    #Mischio l'array per avere tutte immagini diverse
    ax[0].set_title(pixel_dict[columns_names[number]])
    np.random.shuffle(array)
    for col, img_numbr in zip(ax, array[:10]):
        col.imshow(image2imageShaped(img_numbr), cmap = "Greys")
        col.set_axis_off()
    
plt.show()
```
