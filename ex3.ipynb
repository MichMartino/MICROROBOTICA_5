```python
stringa = "Python è un linguaggio potente. Python facilita lo sviluppo rapido!"
punteggiatura = ".,!?:;"

frequenza = {}
parole = stringa.split(" ")
parole = [parola.strip(punteggiatura) for parola in parole]

for parola in parole:
    if parola in frequenza:
        frequenza[parola] += 1
    else:
        frequenza[parola] = 1
print(frequenza)
```
