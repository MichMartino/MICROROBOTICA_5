Capire come funziona l'IA generativa


```python
f = open("./sei_personaggi_in_cerca_dautore_Luigi_Pirandello.txt", "r", encoding="utf-8")
text = f.read()
f.close()
```

```python
text.replace("\n", " ").lower()
for _ in range(3):
    text = text.replace(" ", " ")
# print(text)
```

```python
l_context = 3
generator = {}
for i in range(len(text)- l_context):
    sequenza = text[i:i+l_context]
    if sequenza in generator:
        generator[sequenza].append(text[i+l_context])
    else:
        generator[sequenza] = [text[i+l_context]]
```
