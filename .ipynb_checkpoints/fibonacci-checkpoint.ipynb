```python
def fibonacci_ricorsiva(n):
    #Casi base
    if n == 0:
        return 0
    elif n == 1:
        return 1
    
    #Chiamata ricorsiva
    return fibonacci_ricorsiva(n - 1) + fibonacci_ricorsiva(n - 2)

print("Primi 10 numeri di Fibonacci: ")
for i in range(10):
    print({fibonacci_ricorsiva(i)})
```
