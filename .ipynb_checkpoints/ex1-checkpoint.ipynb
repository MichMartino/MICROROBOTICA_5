```python
def applicasconto(prezzi, soglia, sconto):
    return sorted([(prezzo, prezzo - prezzo * sconto)
           for prezzo in prezzi if prezzo > soglia])
```

```python
prezzi = [25.99, 45.50, 12.30, 67.80, 33.25, 89.99] 
soglia = 30 
sconto = 0.15
print(applicasconto(prezzi, soglia, sconto))
```
