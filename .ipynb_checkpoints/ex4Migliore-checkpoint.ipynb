```python
studenti = {
    "Mario": [("Matematica", 8.5), ("Fisica", 7.0), ("Chimica", 9.0)],
    "Anna": [("Matematica", 9.5), ("Fisica", 8.5), ("Storia", 8.0)],
    "Luca": [("Matematica", 7.0), ("Chimica", 7.5), ("Storia", 9.5)]
}

medie = {}
for nome in studenti:
    voti_studente = studenti[nome]
    somma = 0
    for corso, voto in voti_studente:
        somma = somma + voto
    media = somma / len(voti_studente)
    medie[nome] = media
print(medie)

corsi = {}
for nome in studenti:
    voti_studente = studenti[nome]
    for corso, voto in voti_studente:
        if corso in corsi:
            corsi[corso].append((nome, voto))
        else:
            corsi[corso] = [(nome, voto)]

migliori = {}
for corso in corsi:
    studenti_corso = corsi[corso]
    miglior_voto = 0
    miglior_nome = ""
    for nome, voto in studenti_corso:
        if voto > miglior_voto:
            miglior_voto = voto
            miglior_nome = nome
    migliori[corso] = (miglior_nome, miglior_voto)
print(f"{migliori}")

soglia = 8.5
sopraSoglia = []
for nome in medie:
    if medie[nome] > soglia:
        sopraSoglia.append(nome)
print(sopraSoglia)
```
