# Esercizio 2: Dizionari - Sistema inventario magazzino
Descrizione: Gestisci un inventario con operazioni di aggiunta, rimozione e controllo stock.
Implementa funzioni per aggiornare quantità, verificare disponibilità e mostrare prodotti in
esaurimento.

Input di esempio:

inventario = {"laptop": 15, "mouse": 50, "keyboard": 25, "monitor": 8}

operazioni = [("laptop", -3), ("mouse", 20), ("tablet", 10), ("monitor", -8)]

Output atteso:
Inventario aggiornato: {'laptop': 12, 'mouse': 70, 'keyboard': 25, 'monitor': 0,
'tablet': 10}

Prodotti in esaurimento: ['monitor']

```python
tupla = (1, 2, 3, 4)
print(tupla[2])
```

```python
def verifica_disp(inventario, articolo):
    return inventario[articolo]

def esaurimento(inventario):
    return [articolo for articolo in inventario if inventario[articolo] == 0]

def aggiorna(inventario, operazioni):
    for operazione in operazioni:
        articolo, variazione = operazione
        if articolo in inventario:
            inventario[articolo] = inventario[articolo] + variazione
        else:
            print(f"L'articolo {articolo} non è presente")
            inventario[articolo] = 0
    return inventario
```

```python
inventario = {"laptop": 15, "mouse": 50, "keyboard": 25, "monitor": 8, "GPU": 0}
operazioni = [("laptop", -3), ("mouse", 20), ("tablet", 10), ("monitor", -8)]
```

```python
print(verifica_disp(inventario, "mouse"))
print(esaurimento(inventario))
print(aggiorna(inventario, operazioni))
```
