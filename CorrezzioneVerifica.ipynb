```python
libri = [("nome della rosa", 3), ("dune", 2), ("1984", 0)]

operazioni = ["nome della rosa", "1984", "maus"]

diz = {titolo:disp for titolo, disp in libri}

print(diz)
```

```python
def prestiti(diz, operazioni):
    for op in operazioni:
        if op in diz:
            if diz[op] > 0:
                diz[op] -= 1
    libri = [(titolo, disp) for titolo, disp in diz.items()]
    libri_nonDisp = [titolo for titolo, disp in diz.items() if disp == 0]

    return libri, libri_nonDisp
```

```python
prestiti(diz, operazioni)
```
