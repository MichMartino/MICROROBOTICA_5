```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: np.sin(x)**2
g = lambda x: np.exp(-x**2) #calcola e alla x

#è come scrivere:
# def f(x):
#   return np.sin(x)**2"""
```

```python
x = np.linspace(-5, 5, 1000) #Crea un array di valori equispaziati tra -10 e 10
y_f = f(x)
y_g = g(x)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y_f, label = r"$f(x)=(\sin{x})^2$", linewidth = 2.)
ax.plot(x, y_g, label = r"$g(x)=e^{-x^2}$", linewidth = 2., linestyle = "--")
ax.grid()
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.legend(loc="upper right")
plt.show()
```
