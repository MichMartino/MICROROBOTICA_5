## n righe e 2n colonne

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10
scacchiera = np.zeros((N, N))
scacchiera[::2, ::2] = 1
scacchiera[1::2, 1::2] = 1
scacchiera
```

```python
matrice = np.zeros([N, 2*N])
matrice_random = np.random.uniform(0.,1., (N, N))
matrice[:, N:] = scacchiera   #colonne diviso 2
matrice[:, :N] = matrice_random #colonne diviso 2
plt.imshow(matrice, cmap= "Blues")
```
