```python
#EX 2
#Fatto con funzione roll di numpy
def analizzaVettore(array):
    corrente = array
    precedente = np.roll(array, 1)
    if np.all((precedente < corrente)[1:]):
        risposta = 1
    elif np.all((precedente > corrente)[1:]):
        risposta = 2
    else:
        risposta = 3
    return risposta

#Fatto come lo vuole il professore Tos
def analizzaVettore2(array):
    if np.all(array[:-1] < array[1:]):
        return 1
    if np.all(array[:-1] > array[1:]):
        return -1
    return 0
```

```python
#array = np.random.randint(0, 5,(5))
array2 = np.array([4, 3, 2, 1])
array2
```

```python
if analizzaVettore(array2) == 1:
    print("Crescente")
elif analizzaVettore(array2) == 2:
    print("Decrescente")
else:
    print("Nessuno dei due")
```

```python
#EX 8
arrayDeci = np.arange(0., 1., 0.0001)
arrayDeci
```

```python
len(arrayDeci)
```

```python
f1 = 90  #Frequenza prima onda (Hz)
f2 = 100  #Frequenza seconda onda (Hz)
A1 = 1  #Ampiezza prima onda
A2 = 1  #Ampiezza seconda onda

#Calcolo le due onde
onda1 = A1 * np.sin(2 * np.pi * f1 * arrayDeci)
onda2 = A2 * np.sin(2 * np.pi * f2 * arrayDeci)

intensità = onda1 + onda2
print(intensità)
```

```python
len(intensità)
```

```python
#1. punto
ampiezza_max = np.max(intensità)
ampiezza_max
```

```python
soglia = ampiezza_max / 2
soglia
```

```python
istanti_sopra_soglia = arrayDeci[intensità > soglia]
istanti_sopra_soglia
```

```python
#2. punto
precedente = np.roll(intensità, 1)
successivo = np.roll(intensità, -1)

massimi_locali = (intensità > precedente) & (intensità > successivo)
massimi_locali
```

```python
istanti_massimi = arrayDeci[massimi_locali]
istanti_massimi
```

```python
valori_massimi = intensità[massimi_locali]
valori_massimi
```

```python
#3. punto
massimo_assoluto = np.max(intensità)
massimo_assoluto
```

```python
istanti_massimi_assoluti = arrayDeci[intensità == massimo_assoluto]
istanti_massimi_assoluti
```

```python
#4. punto
istanti_zero = arrayDeci[intensità == 0.]
istanti_zero
```

```python
#EX 9
#1. punto
n_lanci = 1000
risultati = []
for i in range(10):
    francesco = np.random.randint(1, 7, n_lanci)
    giovanni = np.random.randint(1, 7, n_lanci)
    #print(f"I lanci di Francesco: {francesco}")
    #print(f"I lanci di Giovanni: {giovanni}")
    francesco_vince = francesco > giovanni
    giovanni_vince = giovanni > francesco
    pareggio = francesco == giovanni

    punti_francesco = np.sum(francesco_vince) * 1 + np.sum(pareggio) * 0.5
    punti_giovanni = np.sum(giovanni_vince) * 1 + np.sum(pareggio) * 0.5
    risultati.append((punti_francesco, punti_giovanni))
risultati
```
