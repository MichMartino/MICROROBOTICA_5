```python
# ES2
matrice2 = np.random.randint(-10, 11, (6,6)) #estremi inclusi, sinistro sempre incluso, destro no
matrice2
```

```python
matrice_negativi = matrice2 < 0
matrice_negativi #matrice booleana
```

```python
matrice2[matrice_negativi] = 0
matrice2
```

```python
somma_righe = np.sum(matrice2, axis=1) #Asse 0 = colonne Asse 1 = righe
somma_righe
```

```python
solo_maggiori = somma_righe > 15
matrice2[solo_maggiori]
```
