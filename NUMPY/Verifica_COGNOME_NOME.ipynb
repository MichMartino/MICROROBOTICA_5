## ESERCIZIO 1
Apri l'immagine `image.jpg` come array di Numpy. Crea una funzione che utilizzi NumPy per creare una immagine compressa nel seguente modo:
- rimuovendo tutte le righe dispari (0 è considerato pari);
- rimuovendo tutte le colonne dispari (0 è considerato pari).

L'immagine ottenuta avrà un numero di righe pari alla metà di quelle iniziali, un numero di colonne pari alla metà di quelle iniziali. La funzione restituisce una deep copy dell'immagine compressa in tal modo.

Crea una figura con 4 sistemi di assi, dove visualizzi:
* l'immagine originale
* l'immagine compressa una volta
* l'immagine compressa due volte
* l'immagine compressa tre volte

**NON PUOI USARE CICLI, IF E LISTE**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def comprimiImmagine(img):
    imgCopy = np.copy(img)
    matrice_temp = imgCopy[0::2, 0::2, :] #prendo solo le pari, tolgo tutte le dispari
    return matrice_temp
```

```python
img = plt.imread("./images.jpeg")
```

```python
img1 = comprimiImmagine(img)
#Ora comprimo tutte le altre volte
img2 = comprimiImmagine(img1)
img3 = comprimiImmagine(img2)
```

```python
fig, axs = plt.subplots(nrows = 2, ncols = 2, figsize = (8,8))
axs[0, 0].imshow(img)
axs[0, 1].imshow(img1)
axs[1, 0].imshow(img2)
axs[1, 1].imshow(img3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# x ha esattamente 500 punti
x = np.linspace(-3, 3, 500)

# y viene inizializzato per avere la stessa dimensione (500 punti)
y = np.exp(x) # y ha 500 punti

# L'indicizzazione booleana 'x > 0' crea una maschera di 500 elementi (True/False)
# Il calcolo 'np.cos(2 * np.pi * x[x > 0])' viene eseguito solo sui subset corretti di x
# e il risultato viene assegnato solo ai punti corrispondenti in y.
y[x > 0] = np.cos(2 * np.pi * x[x > 0])
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(y, linewidth = 2)
ax.grid()
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.legend(loc="upper right")
plt.show()
```
