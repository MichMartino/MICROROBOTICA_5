## Tipi di grafici

* `ax.plot()`: grafici a linea oppure grafici a dispersione in cui i punti hanno tutti lo stesso stile
* `ax.scatter()`: grafici a dispersione

un **grafico a dispersione** è un grafico in cui si mostra con una variabile *y* dipende da una variabile *x*

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N_PUNTI = 100
x = np.random.uniform(0., 1., N_PUNTI)
y = np.random.uniform(0., 1., N_PUNTI)
c = np.random.uniform(0., 1., (N_PUNTI, 3))
s = np.clip(np.random.normal(200, 100, N_PUNTI), 10, 300) #tutti i valori minori di 10 diventano 10, tutti i valori maggiori di 300 diventano 300
alpha = x = np.random.uniform(0., 1., N_PUNTI)
```

```python
fig, ax = plt.subplots(figsize = (3,3))
ax.scatter(x, y, c=c, s=s, alpha=alpha, marker="h", edgecolor="black")
ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.show()
```

```python
N_PUNTI = 100
x = np.random.uniform(0., 1., N_PUNTI)
y = np.random.uniform(0., 1., N_PUNTI)
c = np.random.uniform(-1000., 1000., N_PUNTI)
s = np.clip(np.random.normal(200, 100, N_PUNTI), 10, 300) #tutti i valori minori di 10 diventano 10, tutti i valori maggiori di 300 diventano 300
alpha = x = np.random.uniform(0., 1., N_PUNTI)
```

```python
fig, ax = plt.subplots(figsize = (3,3))
ax.set_facecolor("Black")
ax.scatter(x, y, c=c, s=s, alpha=alpha, marker="h", edgecolor="white", cmap="RdPu")
ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.show()
```

* `ax.hist()`: Istogrammi, da non confondere con i diagrammi a barre

```python
N_ALUNNI = 10000
altezza = np.random.normal(1.70, 0.1, N_ALUNNI)
#altezza = np.random.uniform(1.60, 1.95, N_ALUNNI) non serve a niente per machine learning fatta così
```

```python
fig, ax = plt.subplots(figsize = (3, 3))
ax.hist(altezza, bins = 20) #bins p il numero delle barre
ax.set_xlabel("altezza")
ax.set_ylabel("N_ALUNNI")
ax.grid()
plt.show()
```

```python
#Simulare il lancio di due dadi, il risultato è dato dalla somma
N_LANCI = 1000000

#Fare istogramma dei punteggi ottenuti dopo n_lanci di due dadi a 6 facce, 11 bins
dado1 = np.random.randint(1, 7, N_LANCI)
dado2 = np.random.randint(1, 7, N_LANCI)
punteggi = dado1 + dado2

fig, ax = plt.subplots(figsize = (4, 4))
ax.hist(punteggi, bins = 11, edgecolor = "red") #bins p il numero delle barre, 11 valori massimi per somma tra due dadi da 12 facce
ax.set_xlabel("valori")
ax.set_ylabel("N_LANCI")
ax.grid()
plt.show()
```

## Esercizio 4

Usa numpy per creare l'immagine seguente composta da nRighe e 2n colonne. l'intensità del colore è casuale nella metà sinistra della matrice, a scacchiera nella metà destra.

non puoi usare cicli, if e liste

```python
import numpy as np
import matplotlib.pyplot as plt

N = 10
a = np.zeros((N, N*2))

# metà sinistra (N × N) casuale
a[:, :N] = np.random.rand(N, N)

# metà destra (N × N) a scacchiera
x = np.arange(N).reshape(-1, 1)
a[:, N:] = (x + np.arange(N)) % 2

#plt.show(a)
```
