```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N_LANCI = 10
N_PARTITE = 3

a = np.random.randint(1, 7,(N_PARTITE, N_LANCI))
b = np.random.randint(1, 7, (N_PARTITE, N_LANCI))
```

```python
a_wins = a > b
b_wins = b > a
pareggio = a == b


punteggio_a = 1. * a_wins + 0.5 * pareggio
#punteggio_b = 1. * b_wins + 0.5 * pareggio
totale_a = np.sum(punteggio_a, axis = 1)

print(f"il giocatore A ha totalizzato {totale_a} punti")
print(f"il giocatore B ha totalizzato {N_LANCI - totale_a} punti")

#print(np.sum(punteggio_a, axis = 1))
#print(np.sum(punteggio_b, axis = 1))
```

```python
if np.count_nonzero(totale_a > (N_LANCI // 2)) > N_PARTITE // 2:
    print("Il giocatore a ha vinto")
```
