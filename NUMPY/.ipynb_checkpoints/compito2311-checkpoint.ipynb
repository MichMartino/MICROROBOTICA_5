```python
import numpy as np

#100 punti nel piano
points = np.random.randint(1, 20, (100, 2))

#distanza dall’origine, formula: √(x² + y²)
distances = np.sqrt(np.sum(points**2, axis=1))

distances 
```

```python
import numpy as np

R, C = 6, 6   #dimensione matrice
p = 8         #elementi da inserire

M = np.zeros((R, C))

#estraggo p posizioni distinte
idx = np.random.choice(R * C, p, replace=False)

#trasformo gli indici in (riga, colonna)
rows = idx // C
cols = idx % C

#assegno i valori
M[rows, cols] = 1
print(M)
```
