```python
import numpy as np
```

**Array da lista**

```python
m = np.array([[1, 2, 4],[5, 6, 7]], dtype = np.int32)
print(m.shape)
print(m)
```

### Shape degli array

La shape di un array può essere:

- un numero (array 1d)
- una tupla da un elemento (array 1d come sopra)
- una tupla da due elementi (array 2d = matrice)
- una tupla da tre elementi (array 3d, per esempio immagine RGB o RGBA)

## Array di numeri pseudocasuali

interi pseudocasuali

```python
a = np.random.randint(1, 7, (3, 2)) #La shape è una tupla
a
```

**Float distribuiti in maniera uniforme su un intervallo [a, b]**

```python
a, b = 5., 6
array = np.random.uniform(a, b, (8,))
array
```

**FLoat distribuiti in maniera gaussiana/ maniera normale con media $m$ e deviazione standard $s$**

```python
m, s = 171., 10.
array = np.random.normal(m, s, 5)
array
```

**Scelta casuale di numeri/stringhe a partire da una sequenza (Lista o un array)**

```python
sequenza = ["Testa", "Croce"]
array = np.random.choice(sequenza, (5,2))
array
```

```python
#si può anche scrivere con random.randint
sequenza = np.arange(1, 7)
array = np.random.choice(sequenza, (5, 2))
array
```

```python
sequenza = ["Testa", "Croce"]
array = np.random.choice(sequenza, (5, 2), p = [0.6, 0.4]) #p è la probabilità, moneta truccata
array
```

```python
sequenza = ["Mario", "Luigi", "Giulia", "Bob"]
array = np.random.choice(sequenza, 2, replace = False) #Così non possono uscire due uguali
array
```

**Permutazione di un array in-place**

```python
array = np.array([0., 1., 10.], dtype = np.float32)
print(array)
np.random.shuffle(array) #Permutazione è mischiare l'array, non restituisce
print(array)
```

**Seme del generatore pseudocasuale**

```python
np.random.seed(42)
sequenza = ["Testa", "Croce"]
array = np.random.choice(sequenza, (5,2))
array
```

### Funzioni per generare array di sequenze numeriche

**Sequenze di interi**

```python
array = np.arange(-4, 5) #senza gap
array
```

```python
array = np.arange(-4, 5, 2) #con gap di 2
array
```

**Valori float equispaziati**

```python
array = np.linspace(0., 1., 11)
array
```

## Funzioni per generare array costanti

**array di zeri**

```python
array = np.zeros((3, 4), dtype = np.int16)
array
```

**array di uni**

```python
array = np.ones((3,4), dtype = np.float16)
array
```

# Esercizi

```python
#Immagine quadrata in cui ogni colonna è una graduazione di colore che noi percepiamo da sinistra verso destra
import matplotlib.pyplot as plt
import numpy as np
```

```python
SIZE = (100, 100)
row = np.linspace(0., 1., SIZE[1]) #Questa è una riga cdell'immagine, una graduazione che va da 0 a 1
matrix = np.zeros(SIZE, dtype = np.float32) #Ora devo appiccicare la riga di graduali dentro la matrice
matrix[:] = row #Tutte le righe della matrice devono essere uguali alla riga che ho creato
#matrix[:] = row[::-1] #per farlo da destra verso sinistra
matrix
```

```python
fig, ax = plt.subplots(figsize = (3,3))
ax.imshow(matrix, cmap="Reds")
plt.show()
```

```python
img = plt.imread("./images.jpeg")
img.shape
```

```python
fig, ax = plt.subplots(figsize = (3,3))
ax.imshow(img) #No colormap, usa i suoi colori predefiniti
plt.show()
```

### Più sistemi di assi nella stessa figura

```python
img = img[:, 0:140, :] #Mi taglia l'immagine a metà
```

```python
fig, axs = plt.subplots(nrows = 2, ncols = 2, figsize = (3,3)) #axs è un array numpy di sistemi di assi
axs[0, 0].imshow(img) #è una matrice!
axs[0, 1].imshow(img[:, :, 0], cmap = "Reds") #canale RED, img[:, :, 0] segue l'ordine..
#righe, colonne e canali di colore, R g b, red è il primo, cioè 0
axs[1, 0].imshow(img[:, :, 1], cmap = "Greens")
axs[1, 1].imshow(img[:, :, 2], cmap = "Blues")
plt.show()
```

```python
#Immagine casuale di valori casuali, gaussiani
img_random_normal = np.clip(np.random.normal(0.5, 1., (100, 100, 3)), 0., 1.)
fig, ax = plt.subplots(figsize = (3,3))
ax.imshow(img_random_normal)
plt.show()
```
