```python
#Esercizio lancio di dadi, 1000 lanci di due dadi a sei facce
#Usare una matrice, non due array
#Quante volte la somma dei due dadi è esattamente 7?
#Quante volte i due dadi mostrano lo stesso valore?
#Media del punteggio ottenuto
#moda delle somma (Il valore che appare con la maggiore frequenza)
```

```python
import numpy as np

N_LANCI = 1000
N_DADI = 2
N_FACCE = 6

matrice_dadi = np.random.randint(1, N_FACCE + 1, (N_DADI, N_LANCI))
print(matrice_dadi)
```

```python
#Indicizzazione booleana per trovare il numero dei 7
matrice_temporanea = matrice_dadi[0] + matrice_dadi[1] == 7
numeri_volte_7 = np.count_nonzero(matrice_temporanea)
print(numeri_volte_7)
```

```python
#Stessa cosa per valori uguali
matrice_temporanea = matrice_dadi[0] == matrice_dadi[1]
#print(matrice_temporanea)
numero_volte_stesso_valore = np.count_nonzero(matrice_temporanea)
print(numero_volte_stesso_valore)
```

```python
#Media del punteggio ottenuto TOTALE
punteggio_temporaneo = np.sum(matrice_dadi)
#print(punteggio_temporaneo)
media = punteggio_temporaneo / N_LANCI
print(media)

#metodo_veloce per medie totale
somme = np.sum(matrice_dadi, axis = 0)
medie = np.mean(somme)
print(medie)

#Media del punteggio per ogni soma
punteggio_somme = matrice_dadi[0] + matrice_dadi[1]
medie = punteggio_somme / N_DADI
print(medie)
```

```python
#Moda delle somme
matrice_temporanea = matrice_dadi[0] + matrice_dadi[1]
print(matrice_temporanea)
moda_somme = np.argmax(np.bincount(matrice_temporanea))
print(np.bincount(matrice_temporanea)) #mi conta quante volte esce quella somma
print(moda_somme) #è per forza 7
```
