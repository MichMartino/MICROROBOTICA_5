```python
import numpy as np

#Fare un sorteggiatore di alunni per interrogarli
N_ALUNNI = 10
#Sorteggiare un alunno da essere interrogato, tutti gli alunni hanno la stessa probabilità, tranne uno che ha 
#probabilità doppia
```

```python
alunni = np.arange(1, N_ALUNNI + 1)
print(alunni)
prob_truccato = [1/11, 2/11, 1/11, 1/11, 1/11, 1/11, 1/11, 1/11, 1/11, 1/11]
n = 8

p = np.ones(len(alunni)) * 1/11 #Probabilità truccate migliori
p[0] = 2/11
print(p)

arrayInterrogati = np.random.choice(alunni, n, replace = False, p = p) #p è la probabilità, moneta truccata
print(arrayInterrogati)
```
