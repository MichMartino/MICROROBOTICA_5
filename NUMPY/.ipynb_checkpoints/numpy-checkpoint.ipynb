```python
n_lanci = 5
alice = np.random.randint(1, 7, n_lanci) #L'ultima variabile è il numero di lanci
bob = np.random.randint(1, 7, n_lanci)
print(f"I lanci di Alice: {alice}")
print(f"I lanci di Bob: {bob}")
```

```python
alice_wins = alice > bob #Solo se i due array sono lunghi uguali, si usa broadcast per due shape uguali..o shape compatibility
alice_wins
```

```python
alice_wins.dtype
```

```python
np.count_nonzero(alice_wins)
```

```python
alice > 4 #Viene spalmato il 4 su tutto l'array, gli array quindi sono "compatibili"
```

```python
img_file = plt.imread("./images.jpeg")
img = img_file.copy() #deepcopy!
```
