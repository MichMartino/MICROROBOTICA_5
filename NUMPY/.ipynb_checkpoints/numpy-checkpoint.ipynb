## Array 1d e funzioni

creiamo il grafico della funzione $ln(x)$

```python
#Generiamo un arrai di valori equispaziati in un intervallo
x = np.linspace(0., 1., 1000) #Estremo sinistro, estremo destro e punti
y = np.log(x) #Se si fa punto di domanda si ha la documentazione del comando
#Broadcast, per ogni valore di x calcola il ln e lo salva in y
y1 = np.exp(x)
```

```python
fig, ax = plt.subplots(figsize = (4,4))#dimensione del grafico in pollici
ax.plot(x,y, label=r"$y=\ln(x)$", color="black", linewidth=3) #r sta per raw, per il markdown
ax.plot(x,y1, label=r"$y=e^x$", color="red", linewidth=3)
ax.grid()
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
plt.show()
```

```python
plt.plot(x,y)
```

```python
#Generiamo un arrai di valori equispaziati in un intervallo
x = np.linspace(0,4*np.pi, 1000) #Estremo sinistro, estremo destro e punti
y = np.sin(x) #Se si fa punto di domanda si ha la documentazione del comando
#Broadcast, per ogni valore di x calcola il ln e lo salva in y
```

```python
fig, ax = plt.subplots(figsize = (6,3))#dimensione del grafico in pollici
ax.plot(x,y, label=r"$y=sin(x)$", color="black", linewidth=3) #r sta per raw, per il markdown
ax.grid()
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
plt.show()
```

```python
# implementare np.argmax usando np.max() e l'inidicazzione booleana
a = np.array([0, 5, 6, 9, 4, 3])
```

```python
NumMax = np.max(a) #Il numero max
indicizzato = a == NumMax #Mi faccio indicizzazzione booleana
print(indicizzato)
indici = np.arange(0, len(a)) #Faccio un vettore di indici
print(indici)
indice = indici[indicizzato][0] #Prende solo il primo dovessero esserci due massimi uguali
print(indice)
```

```python
a = np.array([1,2,3,4,5,10,20])
a
```

```python
np.roll(a, 1) #Shifta l'array di n, quanto metto dopo il nome dell'array, verso destra
```

```python
np.roll(a, -1) #shifta l'array verso sinistra
```

```python
#Usare np.roll() per verificare se un array 1d è monotono crescente. Vietato usare il for
#ogni elemento deve essere maggiore di quello prima
```

```python
a
```

```python
precedente = np.roll(a, 1)
print(precedente)
```

```python
(precedente < a)[1:] #Valore precedente sempre minore del valore dopo tranne il primo
```

```python
np.all((precedente < a)[1:]) #Gli dai un array di bool e lui fa la AND tra tutti
```
