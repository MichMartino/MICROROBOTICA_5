```python
import numpy as np
import matplotlib.pyplot as plt

# ESERCIZIO 1: Matrice casuale + scacchiera
n = 10
matrice = np.random.rand(n, 2*n)

# Creo scacchiera per metà destra
rows, cols = np.meshgrid(np.arange(n), np.arange(n, 2*n), indexing='ij')
matrice[:, n:] = (rows + cols) % 2

plt.figure(figsize=(10, 4))
plt.imshow(matrice, cmap='Reds')
plt.title('Esercizio 1: Sinistra casuale | Destra scacchiera')
plt.colorbar()
plt.show()
```

```python
RIGHE= 10
COLONNE = RIGHE *2
scacchiera = np.zeros([RIGHE,RIGHE])
scacchiera[::2,::2] = 1
scacchiera[1::2,1::2] = 1
scacchiera
matrice_random= np.random.uniform(0.,1.,(RIGHE,RIGHE))
matrice= np.zeros([RIGHE,COLONNE])
matrice[:,RIGHE:]=scacchiera
matrice[:,:RIGHE] = matrice_random
plt.imshow(matrice, cmap="Reds")
```

```python
# ESERCIZIO 2: Distanze dall'origine
punti = np.random.uniform(-10, 10, size=(100, 5))
distanze = np.linalg.norm(punti, axis=1)
plt.figure(figsize=(8, 8))
plt.scatter(punti[:, 0], punti[:, 1], c=distanze, cmap='viridis')
plt.plot(0, 0, 'r*', markersize=15)
plt.title('Esercizio 2: Punti colorati per distanza dall\'origine')
plt.grid(True)
plt.show()
```

```python
# ESERCIZIO 3: Matrice con elementi sparsi
rows, cols, p = 20, 20, 50
matrice = np.zeros((rows, cols))

# Scelgo p posizioni distinte
indici = np.random.choice(rows * cols, size=p, replace=False)
matrice[indici // cols, indici % cols] = np.random.rand(p)

plt.figure(figsize=(7, 7))
plt.imshow(matrice, cmap='hot')
plt.title(f'Esercizio 3: {p} elementi casuali in posizioni distinte')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Creare e plottare un'immagine RGB di dimensione (8,8)
# I tre canali sono generati in maniera pseudo-casuale gaussiana con media zero

RIGHE = 8
COLONNE = 8

# Generare tre canali con distribuzione gaussiana (media 0)
R = np.random.normal(0, 1, (RIGHE, COLONNE))
G = np.random.normal(0, 1, (RIGHE, COLONNE))
B = np.random.normal(0, 1, (RIGHE, COLONNE))

# Combinare i tre canali in un'immagine RGB
img = np.stack([R, G, B], axis=2)

# Normalizzare i valori tra 0 e 1 per la visualizzazione
img = (img - img.min()) / (img.max() - img.min())

plt.imshow(img)
plt.title("Immagine RGB 8x8 con canali gaussiani")
plt.axis('off')
plt.show()
```
