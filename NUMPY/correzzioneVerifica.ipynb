```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 8
x = np.arange(N) + 1
print(x)
y = np.arange(N)
print(y)
```

```python
matr = np.add.outer(x, y) #add è un insieme di funzioni che fa le combo tra le varie funzioni
print(matr)
```

```python
matr = matr % 2
matr
```

```python
matr = matr.astype(float)
matr
```

```python
plt.imshow(matr, cmap = "grey")
```

```python
#rumore = np.random.uniform(
noise = 0.8+0.2*np.random.rand(N, N)
```

```python
matr = matr*noise
plt.imshow(matr, cmap = "grey")
```
